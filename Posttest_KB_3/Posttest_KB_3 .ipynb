# Menggunakan pandas untuk menjalankan program

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Membaca dataset

```python

df = pd.read_csv("tornados.csv")
```

# Membuat Visualisasi 1: Pie Chart - Kategori Bulan Terjadi Tornado

```python

tornado_counts = df['mo'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(tornado_counts, labels=tornado_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Kategori Bulan Terjadi Tornado')
plt.axis('equal')
plt.show()
```

# Visualisasi 2: Heatmap - Korelasi antara variabel numerik

```python

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi Antar Variabel Numerik')
plt.show()
```

# Visualisasi 3: Histogram - Distribusi Jalan Terdampak Tornado

```python

plt.figure(figsize=(8, 6))
plt.hist(df['st'], bins=20, edgecolor='k')
plt.title('Distribusi Jalan Terdampak Tornado')
plt.xlabel('Jalan')
plt.ylabel('Frekuensi')
plt.xticks(rotation = 90)
plt.show()
```

# Visualisasi 4: Bar Plot - Jumlah Tornado per Bulan

```python

df['date'] = pd.to_datetime(df['date'])
df['Month'] = df['date'].dt.month
monthly_tornado_counts = df['Month'].value_counts().sort_index()
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.figure(figsize=(10, 6))
sns.barplot(x=months, y=monthly_tornado_counts)
plt.title('Jumlah Tornado per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Tornado')
plt.xticks(rotation=45)
plt.show()
```
